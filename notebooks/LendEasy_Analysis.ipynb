## 2. Интерпретация

NPV отрицательный, PI меньше 1, IRR отрицательная. Значит, за 3 года проект не окупается. Для венчурного инвестора он может быть интересен только при условии масштабирования после третьего года.

## 3. Управление командой

Product Owner отвечает за стратегию и ценность продукта. Team Lead отвечает за техническую реализацию. Scrum Master помогает команде перейти на Scrum, снижает хаос и устраняет препятствия.
